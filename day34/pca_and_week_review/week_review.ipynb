# Part A :- Concept Application

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
data = load_wine()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target)
```

## 1. Logistic Regression (LR)

### Description: 
Linear model used for classification by estimating class probabilities.
### When to use: 
When data is linearly separable and interpretability matters.
### Key hyperparameters: 
C, solver
### Use case:
 Spam detection

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"Logistic Regression Accuracy: {accuracy*100:.2f} %")
```

## 2. Decision Tree (DT)

### Description:
Tree based model that splits data using feature conditions.
### When to use:
When interpretability and non-linearity are important.
### Key hyperparameters: 
max_depth, criterion
### Use case:
 Loan approval prediction

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"Decision Tree Accuracy: {accuracy*100:.2f} %")
```

## 3. Random Forest (RF)

### Description: 
Ensemble of multiple decision trees using bagging.
### When to use: 
When you need strong general performance with low overfitting
### Key hyperparameters:
 n_estimators, max_depth
### Use case: 
Fraud detection

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"Random Forest Accuracy: {accuracy*100:.2f} %")
```

## 4. AdaBoost

### Description:
 Boosting model that improves weak learners sequentially.
### When to use:
 When simple weak learners need improvement.
### Key hyperparameters: 
n_estimators, learning_rate
### Use case: 
Face detection

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=50)
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"AdaBoost Accuracy: {accuracy*100:.2f} %")
```

## 5. XGBoost

### Description: 
Optimized gradient boosting with regularization.
### When to use: 
For high performance structured data tasks.
### Key hyperparameters: 
max_depth, learning_rate
### Use case:
 Kaggle competitions

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"XGBoost Accuracy: {accuracy*100:.2f} %")
```

## 6. LightGBM

### Description:
 Fast gradient boosting using leaf wise growth.
### When to use: 
Large datasets with speed requirements.
### Key hyperparameters: 
num_leaves, learning_rate
### Use case: 
Large-scale classification

```python
from lightgbm import LGBMClassifier
model = LGBMClassifier()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"LightGBM Accuracy: {accuracy*100:.2f} %")
```

## 7. Voting Classifier

### Description: 
Combines multiple models by majority voting.
### When to use:
To improve robustness.
### Key hyperparameters: 
estimators, voting
### Use case:
 Ensemble baseline

```python
estimators = [
    ('lr', LogisticRegression()),
    ('dt', DecisionTreeClassifier(max_depth=4)),
    ('rf', RandomForestClassifier(n_estimators=100)),
    ('ada', AdaBoostClassifier(n_estimators=50)),
    ('xgb', XGBClassifier()),
    ('lgbm', LGBMClassifier())
]
```

```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=estimators, voting='hard')
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)

print(f"Voting Classifier Accuracy: {accuracy*100:.2f} %")
```

## 8. Stacking

### Description:
 Uses predictions of base models as input to meta model.
### When to use: 
When combining diverse models for better accuracy.
### Key hyperparameters:
 estimators, final_estimator
### Use case:
 Advanced ensemble systems


```python
from sklearn.ensemble import StackingClassifier
model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"Stacking Classifier Accuracy: {accuracy*100:.2f} %")
```

## 9. SVM

### Description: 
Finds optimal separating hyperplane.
### When to use: 
Medium sized datasets with clear margins.
### Key hyperparameters:
 C, kernel
### Use case: 
Text classification

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"SVM Accuracy: {accuracy*100:.2f} %")
```

## 10. KNN

### Description:
 Predicts using nearest neighbors.
### When to use:
 Small datasets with local structure.

### Key hyperparameters:
 n_neighbors, metric
### Use case:
 Recommendation systems

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"KNN Accuracy: {accuracy*100:.2f} %")
```

## 11. K-Means

### Description:
 Clustering using centroid based partitioning.
### When to use:
 Unlabeled data with spherical clusters.

### Key hyperparameters:
 n_clusters, init
### Use case: 
Customer segmentation

```python

from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
model.fit(X_train)
y_pred = model.predict(X_test)
```

## 12. DBSCAN

### Description: 
 Density based clustering detecting outliers automatically.
### When to use:  
Irregular cluster shapes.

### Key hyperparameters:
 eps, min_samples
### Use case:
 Noise detection

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.5)
model.fit(X_train)
y_pred = model.fit_predict(X_test)
```

## 13. PCA

### Description: 
Reduces dimensions by maximizing variance.

### When to use: 
High dimensional data before modeling.

### Key hyperparameters: 
n_components, svd_solver
### Use case:
 Feature reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)

model = KMeans(n_clusters=4)
model.fit(X_pca)
y_pred = model.predict(X_pca)

## Visualization
import matplotlib.pyplot as plt
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap='viridis')
plt.title('KMeans Clusters after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```
