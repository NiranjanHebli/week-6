# Part B : Stretch Problem

## Import Libraries

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

## Load and prepare dataset

```python
def load_text_data(categories=None, test_size=0.2, random_state=42):
    """ Load the 20 Newsgroups dataset, split into training and testing sets, and return the data along with target names."""
    data = fetch_20newsgroups(subset='all', categories=categories, remove=('headers','footers','quotes'))
    X_train, X_test, y_train, y_test = train_test_split(
        data.data, data.target, test_size=test_size, random_state=random_state, stratify=data.target
    )
    return X_train, X_test, y_train, y_test, data.target_names
```

 ## Build TF-IDF + Classifier pipeline

```python
def build_pipeline(model='svm', C=1.0):
    """ Build a text classification pipeline with TF-IDF and the specified model."""
    if model.lower() == 'svm':
        clf = LinearSVC(C=C, max_iter=10000)
    elif model.lower() == 'logistic':
        clf = LogisticRegression(max_iter=1000)
    else:
        raise ValueError("Model must be 'svm' or 'logistic'")
    
    pipeline = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', max_df=0.7)),
        ('clf', clf)
    ])
    return pipeline
```

## Train pipeline

```python
def train_model(pipeline, X_train, y_train):
    """ Train the model using the provided pipeline and training data."""
    pipeline.fit(X_train, y_train)
    return pipeline
```

## Evaluate model

```python
def evaluate_model(pipeline, X_test, y_test, target_names):
    """ Evaluate the model and return accuracy and classification report."""
    y_pred = pipeline.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=target_names)
    return acc, report
```

```python
categories = ['alt.atheism', 'comp.graphics', 'sci.space', 'talk.religion.misc']

# Load data
X_train, X_test, y_train, y_test, target_names = load_text_data(categories)

# SVM
svm_pipeline = build_pipeline('svm', C=1.0)
svm_pipeline = train_model(svm_pipeline, X_train, y_train)
svm_acc, svm_report = evaluate_model(svm_pipeline, X_test, y_test, target_names)

print("LinearSVC Accuracy:", svm_acc * 100)
print(svm_report)

# Logistic Regression
lr_pipeline = build_pipeline('logistic')
lr_pipeline = train_model(lr_pipeline, X_train, y_train)
lr_acc, lr_report = evaluate_model(lr_pipeline, X_test, y_test, target_names)

print("Logistic Regression Accuracy:", lr_acc * 100)
print(lr_report)
```
